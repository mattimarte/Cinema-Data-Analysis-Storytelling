The other values are standardize with a simple calculation

```python
def to_base5(value):
    if isinstance(value, str) and '/' in value:
        var1 =  float(value.split('/')[0])
        var2 = float(value.split('/')[1])
        if var2 == 0:
            return var1 # Avoid division by 0
        else:
            res = (var1 * 5) / var2 # Convert to Base 5
            if res > 5:
                return 5
            else:
                return res
                
```

```python
review_df['Updated_Score'] = review_df['review_score'].apply(lambda x: to_base5(x))
review_df.head(10)
```

## Generation of the new clean files

```python
from sqlalchemy import create_engine, text
```

```python
USER = "postgres"
PASSWORD = "admin"
HOST = "localhost"
PORT = "5432"
DBNAME = "moviesdb"

# Create the Connection
engine = create_engine(f'postgresql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DBNAME}')
```

```python
def insert_if_not_exists(df, table_name, unique_col, engine):
    columns = df.columns.tolist()
    col_names = ", ".join(columns)
    col_placeholders = ", ".join([f":{col}" for col in columns])
    
    insert_stmt = f"""
        INSERT INTO {table_name} ({col_names})
        VALUES ({col_placeholders})
        ON CONFLICT ({unique_col}) DO NOTHING
    """
    
    with engine.begin() as conn:
        for _, row in df.iterrows():
            conn.execute(text(insert_stmt), row.to_dict())
```
